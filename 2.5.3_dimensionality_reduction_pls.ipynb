```python
import math
import warnings
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
from sklearn.cross_decomposition import PLSRegression
%matplotlib inline
warnings.filterwarnings(
    action='ignore',
    module='scipy',
    message='^internal gelsd'
)
```

```python
sns.set(style='white', rc={'figure.figsize':(8,5)})
```

```python
n = 1000
p = 10
X = np.random.normal(size=n * p).reshape((n, p))
y = X[:, 0] + 2 * X[:, 1] + np.random.normal(size=n * 1) + 5


corrmat = pd.DataFrame(np.insert(X, 0, y, axis=1)).corr()

sns.heatmap(corrmat, vmax=0.8, square=True, cmap='bwr', center=0)
plt.show()
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)

y_pred = regr.predict(X)
print('R-squared Linear Regression: ', regr.score(X, y))

pls1 = PLSRegression(n_components=3)
pls1.fit(X, y)

y_PLS_pred = pls1.predict(X)
print('R-squared PLS 3 component: ', pls1.score(X, y))

plt.scatter(y_pred, y_PLS_pred)
plt.xlabel('Predicted by original 10 features')
plt.ylabel('Predicted by 3 features')
plt.title('Comparing LR and PLSR predictions')
plt.show()
```

```python
#same as above with fewer components
pls2 = PLSRegression(n_components=1)
pls2.fit(X, y)

y_PLS_pred = pls2.predict(X)
print('R-squared PLS 1 component: ', pls2.score(X, y))
print('R-squared Linear Regression: ', regr.score(X, y))

plt.scatter(y_pred, y_PLS_pred)
plt.xlabel('Predicted by original 10 features')
plt.ylabel('Predicted by 1 features')
plt.title('Comparing LR and PLSR predictions')
plt.show()
```

```python
#same as first with more components
pls3 = PLSRegression(n_components=6)
pls3.fit(X, y)

y_PLS_pred = pls3.predict(X)
print('R-squared PLS 6 component: ', pls3.score(X, y))
print('R-squared Linear Regression: ', regr.score(X, y))

plt.scatter(y_pred, y_PLS_pred)
plt.xlabel('Predicted by original 10 features')
plt.ylabel('Predicted by 6 features')
plt.title('Comparing LR and PLSR predictions')
plt.show()
```
